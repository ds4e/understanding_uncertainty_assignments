## Exam Topics

1. Probability Spaces
    - Outcomes/Sample Space, Events, Probabilities
    - Probability Axioms ($p(\varnothing)=0$; $p(Z)=1$; if $E_1, ..., E_n$ are disjoint, then $p(E_1 \cup E_2 ... \cup E_n) = \sum_{i=1}^n p(E_i)$)

2. Random Variables
    - A random variable $X$ is a mapping from Sample Space to vectors of real numbers
    - The cumulative distribution function (cdf) is $F(x) = p[X \le x] = p(\{z \in Z: X(z) \le x \})$
    - The probability density function (pdf) is $f(x) = F'(x)$, where it exists
    - The probability mass function (pmf) is $m(x) = p(\{X(z)=x\})$
    - Common distributions (there are many worth knowing, but in terms of muscle memory for an exam):
        - Uniform
        - Normal, log-normal
        - Logistic
        - Bernoulli
        - Exponential
        - Poisson

3. Maximum Likelihood Estimation
    - If $(X_1, X_2, ..., X_n)$ are independent random variables, then their joint density satisfies $f(x_1, x_2, ..., x_n) = f_{X_1}(x_1) f_{X_2}(x_2) ... f_{X_n}(x_n)$
    - The likelihood of a set of data is the joint density of observing it, given the parameters $\beta$: $L(\beta) = \prod_{i=1}^n f_{X_i}(x_i; \beta) $
    - The log-likelihood is the log of the likelihood: $\ell(\beta) = \sum_{i=1}^n \log( f_{X_i}(x_i; \beta) ) $
    - The **maximum likelihood estimator** is the parameter choice that maximizes the likelihood. Differentiate $\ell(\beta)$ with respect to $\beta$, set it equal to zero, and solve for $\hat{\beta}_{MLE}$
    - For the models we play with, the MLE usually looks something like a sample average or sample variance, or a transformation of one

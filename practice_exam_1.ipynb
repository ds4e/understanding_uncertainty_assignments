- Right now, the stock price is 10.
- At the end of today, the stock price goes up by 2 with probability $1/2$ and down by 2 with probability $1/2$
- At the end of tomorrow, the stock price the stock price goes up by 2 with proability $1/2$ and down by 2 with probability $1/2$.

1. Write out the outcomes, events, and probabilties for the stock price at the end of day two.
2. Compute and sketch the cumulative distribution function of the stock price at the end of day two.
3. Compute the expectation and the variance of the stock price at the end of day two.

### 2. Consider the following scenario:

We have three options, $a$, $b$, and $c$. Each option has a value associated with it, $v_a > v_b > v_c$. We want to pick an option at random, or model someone doing so. The **softmax** mass function is
$$
pr[a] = \frac{e^{v_a/\sigma}}{e^{v_a/\sigma} + e^{v_b/\sigma} + e^{v_c/\sigma}}, \quad pr[b] = \frac{e^{v_b/\sigma}}{e^{v_a/\sigma} + e^{v_b/\sigma} + e^{v_c/\sigma}}, \quad pr[c] = \frac{e^{v_c/\sigma}}{e^{v_a/\sigma} + e^{v_b/\sigma} + e^{v_c/\sigma}}
$$

1. Show that this probability mass function over the labels $a$, $b$, and $c$ satisfies the axioms of probability.

2. If we remove alternative $c$, there's an easy way to update the choice probabilities:
$$
pr[a] = \frac{e^{v_a/\sigma}}{e^{v_a/\sigma} + e^{v_b/\sigma}}, \quad pr[b] = \frac{e^{v_b/\sigma}}{e^{v_a/\sigma} + e^{v_b/\sigma} }
$$
Show that regardless of whether option $c$ is available, $pr[a]/pr[b]$ is the same (we often call this **independence of irrelevant alternatives**).

3. Show that as $v_a$ increases, the probability of selecting $a$ goes up while the probabilities of selecting $b$ and $c$ go down.

4. Show that as $\sigma$ gets large ($\sigma \rightarrow \infty$), these probabilities become uniform (i.e. they each take the value 1/3). Show that as $\sigma$ gets small ($\sigma \rightarrow 0$), softmax selects the option with the highest value (this is why it is called "softmax"; hint: divide everything by $e^{v_a/\sigma}$ first). 

### 3. Suppose that $X$ is a random variable with the Poisson distribution, so that 
$$
pr[X=k] = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, ...
$$

1. Show that the Poisson mass function obeys the laws of probability* over $k = 0, 1, 2, ... $
2. Derive the likelihood function for a sample taking values $x_1 = 1$, $x_2 = 2$, and $x_3 = 3$.
3. Maximize the likelihood with respect to $\lambda$ and derive the maximum likelihood estimator. 
4. Using your fitted model, predict the probability that $X \le 2$.

* Hint: $ \sum_{k=0}^\infty \frac{\lambda^k}{k!} = e^{\lambda} $

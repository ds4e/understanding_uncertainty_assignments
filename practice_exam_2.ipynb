# Understanding Uncertainty Exam

### 1. You are working for a hospital, concerned with capacity.

Whenever someone visits the Emergency Room (ER),
- With probability $4/5$ they leave that day
- With probability $1/5$, they are admitted and stay an amount of time that is uniformly distributed between 0 and 30 days. Specifically, if someone is admitted, the cumulative distribution of bed-days at the hospital is described by
$$
F_T(t) = \begin{cases}
0, & t<0 \\
\frac{t}{30}, & 0 \le t \le 30 \\
1, & t>30.
\end{cases}
$$

1. Sketch the distribution function for overll bed-days, including the randomness over whether a patient is admitted or not.
2. Imagine someone is admitted. Compute the expected number of bed-days, and the variance.
3. Now imagine somene has arrived at the ER, but the determination of whether they'll be admitted has not yet been made. Compute the expected number of bed-days.
4. If $n$ patients arrive tomorrow, what is the expected number admitted? What is the expected total number of bed-days needed?

### 2. And now, the weather.

There are two states: Clear and Rainy. The weather evolves as a Markov Chain with transition matrix
$$
\left[ \begin{array}{c|cc} \hline \text{clear} & 	ext{clear} & 	ext{rainy} \ & .6 & .4 \\ \text{rainy}.4 & .6 \end{array} \right]
$$
Recall, $p_{i \leftarrow j}$ is the probability of going from state $j$ to state $i$.

1. If today is clear, what is your forecast for tomorrow?
2. If today is clear, what is your forecast for the day after tomorrow?
3. It today is clear, what is your forecast for the third day?
4. Show that if your current forecast is $(.5, .5)$, then the next forecast is also $(.5, .5)$.

### 3. Suppose there are three labels: $a$, $b$, and $c$.

In our sample, we observe $a$ once, $b$ twice, and $c$ once. We model this as a categorical variable, with probabilities $p_a$, $p_b$, and $p_c$ for each label.

1. What rules must $p_a$, $p_b$, and $p_c$ satisfy in order to obey the axioms of probability over $\{ a, b, c\}$?
2. What is the likelihood for these data? (Hint: Recall the Bernoulli distribution and generalize it to three outcomes.)
3. What is the maximum likelihood estimator for the probabilities $p_a$, $p_b$, and $p_c$? (Hint: $p_c = 1- p_a - p_b$.)
